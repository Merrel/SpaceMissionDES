### Demo Sampling Distributions

```python
from scipy.stats import beta, expon, weibull_min
import pandas as pd
import altair as alt
```

```python
N = 1000
# X = beta(a := 2, b := 5)
# X = expon(loc=0, scale=1)
X = weibull_min(c=0.5, loc=0, scale=0.1)

base_duration = 0

duration = [
    # base_duration + X.rvs()
    X.rvs()
    for _ in range(N)
]

df = pd.DataFrame({"duration": duration})

alt.Chart(df).mark_bar().encode(
    # alt.X("duration:Q", bin=True),
    alt.X("duration:Q", bin=alt.Bin(extent=[0, 1], step=0.05)),
    # alt.X("duration:Q", bin=alt.Bin(extent=[0, 40], step=2)),
    y='count()',
).properties(
    width = 800
)
```
